```python
s = Signal(f0=16000, f1=19000, fs=100000, T=0.0205, A=1000)
```

```python
wu1 = s.chirp_cos() + s.chirp_sin() + s.white_noise(A=7000)
wu2 = s.chirp_cos() + s.chirp_sin() + s.white_noise(A=7000)
wd1 = s.chirp_cos(updown="down") + s.chirp_sin(updown="down") + s.white_noise(A=7000)
wd2 = s.chirp_cos(updown="down") + s.chirp_sin(updown="down") + s.white_noise(A=7000)
```

```python
s.plot_fft(wu1)
```

```python
s.plot_fft(wu1 * s.chirp())
```

```python
s.plot_fft((wu1 + wu2) * s.chirp())
```

```python
s.plot_fft((wu1 + wd1) * s.chirp(updown="down"))
```

```python
s.plot_fft((wd1 + wd2) * s.chirp(updown="down"))
```
