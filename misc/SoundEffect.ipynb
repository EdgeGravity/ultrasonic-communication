```python
%matplotlib inline

import numpy as np
from numpy import zeros, append, linspace, cos, sin, pi, int16, log10
from scipy.io.wavfile import write
from scipy.signal import spectrogram
from scipy.fftpack import fft, ifft, fftfreq, fftshift
import matplotlib.pyplot as plt
import peakutils
from IPython.display import display, Audio
```

```python
SAMPLING_RATE = 44100  # Hz
TIME_FRAME = 2.0 # Time frame in sec
AMPLITUDE = 10000 # 16bit PCM max amplitude
F = 440.0
WAVE_FILE="./Tone.wav"
```

```python
# Tone generator
def generate(f=F, amp=AMPLITUDE, T=TIME_FRAME):
    t = linspace(0, T, int(T * SAMPLING_RATE))
    arg = (2.0 * pi * f * t)
    return sin(arg) * amp
```

```python
tone = generate(f=F)

def play(wave):
    write(WAVE_FILE, SAMPLING_RATE, wave.astype(int16))
    display(Audio('./' + WAVE_FILE))

play(tone)
```

```python
THRES = 500.0
def distortion(s):
    if abs(s) >= THRES:
        return np.sign(s) * THRES
    else:
        return s

tone = generate()
v_distortion = np.vectorize(distortion)
tone = v_distortion(tone)
play(tone*5)
```
