# Raw PCM file analyzer

```python
import pandas
from scipy.io.wavfile import write
from scipy.signal import spectrogram, welch
from numpy import int16, int32
import matplotlib.pyplot as plt

SAMPLING_RATE = 100000
```

```python
df = pandas.read_csv("../agent/out.raw")
WAV_FILE = 'dfsdm_raw.wav'
#print(df["Amplitude"])
write(WAV_FILE, SAMPLING_RATE, df["Amplitude"].astype(int16))
```

```python
f, Pxx = welch(df["Amplitude"], fs=SAMPLING_RATE)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Power')
plt.plot(f, Pxx)
```

```python
plt.figure(figsize=(13, 4), dpi=80)
f, t, Sxx = spectrogram(df["Amplitude"], fs=SAMPLING_RATE)
plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
