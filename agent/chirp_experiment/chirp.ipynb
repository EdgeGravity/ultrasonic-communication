Repetetive frequency-sweep tone (20msec) is transmitted from mini speaker to MEMS mic.

The speaker is a very cheap one.

```python
import matplotlib.pyplot as plt
import pandas
```

## Frequency sweep range: 16000Hz - 16000Hz

```python
df = pandas.read_csv("./16000_48.1(kHz)_M1.fft")
plt.figure(figsize=(13, 4), dpi=80)
df.plot(x="Frequency(Hz)", y="Magnitude(dB)")
df.plot(x="Frequency(Hz)", y="Magnitude")
```

## Noise from paper rubbing

## Conclusion

- Frequency-sweep range 16000Hz - 17000Hz will be suitable, if a very cheap speaker is used to transmit the signal tone. 
- The chirp tone generates a lot of noises of lower frequencies less than 16000Hz. 
- Paper rubbing generates a lot of noise over the frequency-sweep range.
