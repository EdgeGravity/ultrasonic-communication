```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import glob
fft_files = glob.glob('./*.fft')
raw_files = glob.glob('./*.raw')
filter_files = glob.glob('./*.flt')

dfs_fft = []
dfs_raw = []
dfs_filter = []
```

```python
for file in fft_files:
    dfs_fft.append((file.split('_')[2].split('.')[0], pd.read_csv(file)))
#print(dfs_fft)
```

```python
for file in raw_files:
    dfs_raw.append((file.split('_')[2].split('.')[0], pd.read_csv(file)))
#print(dfs_raw)
```

```python
for file in filter_files:
    dfs_filter.append((file.split('_')[2].split('.')[0], pd.read_csv(file)))
#print(dfs_filter)
```

## FFT

```python
for d in dfs_fft:
    mic = d[0]
    df = d[1]
    df.plot(title=mic, x='Frequency(Hz)', y='Magnitude')
```

## Raw data

```python
for d in dfs_raw:
    mic = d[0]
    df = d[1]
    df.plot(title=mic, x='Index', y='Amplitude')
```

## Filtered data (Hanning window)

```python
for d in dfs_filter:
    mic = d[0]
    df = d[1]
    df.plot(title=mic, x='Index', y='Amplitude')
```
