```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
DECIMATION = 5
```

```python
df = pd.read_csv('./1526445492_41.7(kHz)_M1.fft')
df.plot(x='Frequency(Hz)', y='Magnitude')
```

```python
df[0:100].mean()
```

```python
m = int(df.shape[0]/DECIMATION)
dfm = df[0:DECIMATION].mean()
for i in range(1, m):
    j = i * DECIMATION
    dfm = pd.concat([dfm, df[j:j+100].mean()], axis=1, ignore_index=True)
df2 = dfm.T
```

```python
df2.plot(x='Frequency(Hz)', y='Magnitude')
```
